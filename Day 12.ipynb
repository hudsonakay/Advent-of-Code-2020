```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel('input_12.xlsx')
```

```python
data.head()
```

```python
commands = []
for i in range(len(data['Column1'])):
    commands.append(data['Column1'][i])
```

```python
commands
```

```python
start_pos = [0,0]
heading = 'E'
headings = ['N','E','S','W']
for command in commands:
    direction = command[0]
    units = int(command[1:])
    if direction == 'S':
        start_pos[0] -= units
    elif direction == 'N':
        start_pos[0] += units
    elif direction == 'E':
        start_pos[1] += units
    elif direction == 'W':
        start_pos[1] -= units
    elif direction == 'F':
        if heading == 'N':
            start_pos[0] += units
        elif heading == 'E':
            start_pos[1] += units
        elif heading == 'S':
            start_pos[0] -= units
        elif heading == 'W':
            start_pos[1] -= units
    elif direction == 'R':
        heading_index = headings.index(heading)
        pos = int(units/90)
        heading = headings[(heading_index+pos)%4]
    elif direction == 'L':
        heading_index = headings.index(heading)
        pos = int(units/90)
        heading = headings[(heading_index-pos)%4]
    print(start_pos,heading)
```

```python
start_pos = [0,0]
waypoint = [1,10]
for command in commands:
    direction = command[0]
    units = int(command[1:])
    if direction == 'S':
        waypoint[0] -= units
    elif direction == 'N':
        waypoint[0] += units
    elif direction == 'E':
        waypoint[1] += units
    elif direction == 'W':
        waypoint[1] -= units
    elif direction == 'F':
        ns = waypoint[0]
        ew = waypoint[1]
        start_pos[0] += int(ns*units)
        start_pos[1] += int(ew*units)
    elif direction == 'R':
        ns = waypoint[0]
        ew = waypoint[1]
        rot = int(units/90)
        if rot == 1:
            waypoint[0] = -ew
            waypoint[1] = ns
        elif rot == 2:
            waypoint[0] = -ns
            waypoint[1] = -ew
        elif rot == 3:
            waypoint[0] = ew
            waypoint[1] = -ns
    elif direction == 'L':
        ns = waypoint[0]
        ew = waypoint[1]
        rot = int(units/90)
        if rot == 1:
            waypoint[0] = ew
            waypoint[1] = -ns
        elif rot == 2:
            waypoint[0] = -ns
            waypoint[1] = -ew
        elif rot == 3:
            waypoint[0] = -ew
            waypoint[1] = ns
    print(command,start_pos,waypoint)
```
