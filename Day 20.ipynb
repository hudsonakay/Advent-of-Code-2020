```python
def make_puzzle(puzzle,column):
    display_ids = []
    for row in puzzle:
        display_ids.append(row[column])
    display = []
    for tile in display_ids:
        if tile == 0:
            display.append(tile_0)
        else:
            display.append(full_tile_dict[tile])
    return(display)
```

```python
def rotate_90(tile):
    new_rows = []
    for i in range(len(tile[0])):
        new_row = ''
        for row in tile:
            new_row += row[i]
        new_rows.append(new_row[::-1])
    return(new_rows)

def flip_x(tile):
    new_rows = tile[::-1]
    return(new_rows)

def flip_y(tile):
    new_rows = []
    for row in tile:
        new_rows.append(row[::-1])
    return(new_rows)
        
def rotate_tile(tile_id,commands):
    tile = full_tile_dict[tile_id]
    new_tile = []
    for command in commands:
        if command == '90':
            new_tile = rotate_90(tile)
        elif command == '180':
            new_tile = rotate_90(rotate_90(tile))
        elif (command == '270') or (command == 'cc90'):
            new_tile = rotate_90(rotate_90(rotate_90(tile)))
        elif command == 'flipx':
            new_tile = flip_x(tile)
        elif command == 'flipy':
            new_tile = flip_y(tile)
        elif command == 'flipxy':
            new_tile = flip_x(flip_y(tile))
        elif command == 'cc90 and flipy':
            new_tile = flip_y(rotate_90(rotate_90(rotate_90(tile))))
    return(new_tile)
```

```python
def get_left(tile):
    left = ''
    for line in tile:
        left += line[0]
    left = left[::-1]
    return(left)

def get_right(tile_id):
    tile = full_tile_dict[tile_id]
    right = ''
    for line in tile:
        right += line[-1]
    return(right[::-1])

options = ['90','180','270','flipx','flipy','flipxy','cc90 and flipy']

# def find_orientation(good_tile_id,left_tile_right_side):
#     tile = full_tile_dict[good_tile_id]
#     attempt_tile = rotate_tile(good_tile_id,['90'])
#     if get_left()
```
