```python
def dectobin(x):
    num = int(x)
    binary = ['0' for i in range(36)]
    for i in range(36):
        if (num - 2**(35-i)) < 0:
            binary[i] = '0'
        else:
            binary[i] = '1'
            num = (num - 2**(35-i))
    return(''.join(binary))

def bintodec(x):
    dec = 0
    for i in range(36):
        if x[i] == '1':
            dec += 2**(35-i)
    return(dec)
```

```python
def xperms(num):
    num_perms = 2**num
    columns = []
    half = int(num_perms/2)
    multiple = 1
    while half >= 1:
        col = ''
        for i in range(half):
            col += '0'
        for i in range(half):
            col += '1'
        col *= multiple
        columns.append(col)
        multiple *= 2
        half = int(half/2)
    perms = []
    for i in range(num_perms):
        perm = ''
        for j in columns:
            perm += j[i]
        perms.append(perm)
    return(perms)
```

```python
def replace(y,permutations):
    replaced = []
    for perm in permutations:
        replacement = ''
        X_num = 0
        for i in range(36):
            if y[i] == 'X':
                replacement += perm[X_num]
                X_num += 1
            else:
                replacement += y[i]
        replaced.append(replacement)
    return(replaced)
```

```python
def addresses(y):
    X_indices = []
    for i in range(36):
        if y[i] == 'X':
            X_indices.append(i)
    num = len(X_indices)
    permutations = xperms(num)
    replaced = replace(y,permutations)          
    values = []
    for j in replaced:
        values.append(bintodec(j))
    return(values)
```

```python
mem = {}
mask = []
for i in commands:
    if i[0] == 'mask':
        mask = i[1]
    else:
        loc = dectobin(int(i[0][4:-1]))
        storage = int(i[1])
        loc_changed = ''
        for i in range(36):
            if mask[i] == 'X':
                loc_changed += 'X'
            elif mask[i] == '1':
                loc_changed += '1'
            elif mask[i] == '0':
                loc_changed += loc[i]
        address_list = addresses(loc_changed)
        for j in address_list:
            mem[j] = storage
```
