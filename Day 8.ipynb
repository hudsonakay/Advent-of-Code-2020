```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel('input_8.xlsx')
```

```python
data.head()
```

```python
command = []
for i in range(len(data['Column1'])):
    command.append(data['Column1'][i])
```

```python
next_line = []
for i in range(len(data['Column2'])):
    next_line.append(data['Column2'][i])
```

```python
len(next_line)
```

```python
accumulator = 0
used_lines = []
current_line = 0
step = 0
while current_line not in used_lines:
    used_lines.append(current_line)
    comm = command[current_line]
    action = next_line[current_line]
    if comm == 'acc':
        accumulator += action
        current_line += 1
    elif comm == 'jmp':
        current_line += action
    elif comm == 'nop':
        current_line += 1
    step += 1
```

```python
switcher = []
for i in range(len(command)):
    if command[i] == 'jmp':
        switcher.append('nop')
    elif command[i] == 'nop':
        switcher.append('jmp')
    elif command[i] == 'acc':
        switcher.append('acc')
```

```python
for i in range(len(command)):
    accumulator = 0
    used_lines = []
    current_line = 0
    step = 0
    new_command = command.copy()
    new_command[i] = switcher[i]
    while current_line not in used_lines:
        used_lines.append(current_line)
        comm = new_command[current_line]
        action = next_line[current_line]
        if comm == 'acc':
            accumulator += action
            current_line += 1
        elif comm == 'jmp':
            current_line += action
        elif comm == 'nop':
            current_line += 1
        step += 1
        if current_line == 635:
            print(i,accumulator)
            current_line = 0
```
