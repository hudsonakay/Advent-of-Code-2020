```python
data_1 = []
x = ''
for i in range(len(data['Column1'])):
    if pd.isnull(data['Column1'][i]) == False:
        x += data['Column1'][i] + ' '
    else:
        data_1.append(x)
        x = ''
data_1.append(x)
```

```python
def checkbyr(x):
    x = x[4:]
    if len(x) != 4:
        return(False)
    x = int(x)
    if 1920 <= x <= 2002:
        return(True)
    else:
        return(False)
checkbyr('byr:2000')
```

```python
def checkiyr(x):
    x = x[4:]
    if len(x) != 4:
        return(False)
    x = int(x)
    if 2010 <= x <= 2020:
        return(True)
    else:
        return(False)
```

```python
def checkeyr(x):
    x = x[4:]
    if len(x) != 4:
        return(False)
    x = int(x)
    if 2020 <= x <= 2030:
        return(True)
    else:
        return(False)
```

```python
def checkhgt(x):
    x = x[4:]
    if (x[-2:] == 'cm'):
        x = int(x[:-2])
        if 150 <= x <= 193:
            return(True)
        else:
            return(False)
    elif (x[-2:] == 'in'):
        x = int(x[:-2])
        if 59 <= x <= 76:
            return(True)
        else:
            return(False)
    else:
        return(False)
checkhgt('hgt:59in')
```

```python
def checkhcl(x):
    good = ['1','2','3','4','5','6','7','8','9','0','a','b','c','d','e','f']
    x = x[4:]
    if x[0] != '#':
        return(False)
    x = x[1:]
    if len(x) != 6:
        return(False)
    for i in range(6):
        if x[i] not in good:
            return(False)
        else:
            continue
    return(True)
```

```python
def checkecl(x):
    good = ['amb','blu','brn','gry','grn','hzl','oth']
    x = x[4:]
    if len(x) != 3:
        return(False)
    if x in good:
        return(True)
    else:
        return(False)
```

```python
def checkpid(x):
    good = ['1','2','3','4','5','6','7','8','9','0']
    x = x[4:]
    if len(x) != 9:
        return(False)
    for i in range(9):
        if x[i] not in good:
            return(False)
        else:
            continue
    return(True)
```

```python
data_4 = []
for i in range(len(data_3)):
    new = []
    for j in range(len(data_3[i])):
        info = data_3[i][j]
        if 'byr' in info:
            if checkbyr(info) == True:
                new.append(info)
        elif 'iyr' in info:
            if checkiyr(info) == True:
                new.append(info)
        elif 'eyr' in info:
            if checkeyr(info) == True:
                new.append(info)
        elif 'hgt' in info:
            if checkhgt(info) == True:
                new.append(info)
        elif 'hcl' in info:
            if checkhcl(info) == True:
                new.append(info)
        elif 'ecl' in info:
            if checkecl(info) == True:
                new.append(info)
        elif 'pid' in info:
            if checkpid(info) == True:
                new.append(info)
    data_4.append(new)
```
