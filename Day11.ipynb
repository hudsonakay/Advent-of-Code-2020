```python
seats = []
for i in range(len(data['Column1'])):
    seats.append(data['Column1'][i])
empty_row = ''
for i in range(len(data['Column1'][0])):
    empty_row += '.'
seats.insert(0,empty_row)
seats.append(empty_row)
seats_new = []
for i in seats:
    seats_new.append('.' + i + '.')
seats = seats_new

seat_cat = {}
for i in range(len(seats)):
    for j in range(len(seats[i])):
        seat_cat[i,j] = seats[i][j]
```

```python
def fill_seat(x):
    row = x[0]
    col = x[1]
    seats_around = [seat_cat[(row-1,col-1)],seat_cat[(row-1,col)],seat_cat[(row-1,col+1)],seat_cat[(row,col-1)],
                    seat_cat[(row,col+1)],seat_cat[(row+1,col-1)],seat_cat[(row+1,col)],seat_cat[(row+1,col+1)]]
    filled_count = 0
    for i in seats_around:
        if i == '#':
            filled_count += 1
    if (filled_count == 0) and (seat_cat[x] == 'L'):
        return('#')
    elif (filled_count > 3) and (seat_cat[x] == '#'):
        return('L')
    else:
        return(seat_cat[x])
```

```python
for i in range(1000):
    new_seat_cat = seat_cat.copy()
    for j in keys_list:
        if seat_cat[j] == '.':
            new_seat_cat[j] = '.'
        else:
            new_seat_cat[j] = fill_seat(j)
    total_filled = 0
    seat_cat = new_seat_cat.copy()
    for value in seat_cat.values():
        if value == '#':
            total_filled += 1
    print(total_filled)
```

```python
seats = []
for i in range(len(data['Column1'])):
    seats.append(data['Column1'][i])
empty_row = ''
for i in range(len(data['Column1'][0])):
    empty_row += '0'
seats.insert(0,empty_row)
seats.append(empty_row)
seats_new = []
for i in seats:
    seats_new.append('0' + i + '0')
seats = seats_new

seat_cat = {}
for i in range(len(seats)):
    for j in range(len(seats[i])):
        seat_cat[i,j] = seats[i][j]
```

```python
def fill_seat(x):
    row = x[0]
    col = x[1]
    seats_around = [seat_cat[(row-1,col-1)],seat_cat[(row-1,col)],seat_cat[(row-1,col+1)],seat_cat[(row,col-1)],
                    seat_cat[(row,col+1)],seat_cat[(row+1,col-1)],seat_cat[(row+1,col)],seat_cat[(row+1,col+1)]]
    blank = 1
    while seats_around[0] == '.':
        blank += 1
        seats_around[0] = seat_cat[(row-blank,col-blank)]
    blank = 1
    while seats_around[1] == '.':
        blank += 1
        seats_around[1] = seat_cat[(row-blank,col)]
    blank = 1
    while seats_around[2] == '.':
        blank += 1
        seats_around[2] = seat_cat[(row-blank,col+blank)]
    blank = 1
    while seats_around[3] == '.':
        blank += 1
        seats_around[3] = seat_cat[(row,col-blank)]
    blank = 1
    while seats_around[4] == '.':
        blank += 1
        seats_around[4] = seat_cat[(row,col+blank)]
    blank = 1
    while seats_around[5] == '.':
        blank += 1
        seats_around[5] = seat_cat[(row+blank,col-blank)]
    blank = 1
    while seats_around[6] == '.':
        blank += 1
        seats_around[6] = seat_cat[(row+blank,col)]
    blank = 1
    while seats_around[7] == '.':
        blank += 1
        seats_around[7] = seat_cat[(row+blank,col+blank)]
    filled_count = 0
    for i in seats_around:
        if i == '#':
            filled_count += 1
    if (filled_count == 0) and (seat_cat[x] == 'L'):
        return('#')
    elif (filled_count > 4) and (seat_cat[x] == '#'):
        return('L')
    else:
        return(seat_cat[x])
```

```python
for i in range(1000):
    new_seat_cat = seat_cat.copy()
    for j in keys_list:
        if seat_cat[j] == '.':
            new_seat_cat[j] = '.'
        elif seat_cat[j] == '0':
            new_seat_cat[j] = '0'
        else:
            new_seat_cat[j] = fill_seat(j)
    total_filled = 0
    seat_cat = new_seat_cat.copy()
    for value in seat_cat.values():
        if value == '#':
            total_filled += 1
    print(total_filled)
```
